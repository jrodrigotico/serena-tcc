```python
import requests
import json
import pandas as pd
from datetime import datetime
from dateutil.relativedelta import relativedelta
import pytz

api_key = '45bd4693a487a8ad3a2a2dd40a1bc30d'

df_total = pd.DataFrame()

# Loop de janeiro/2023 a dezembro/2024
start = datetime(2023, 1, 1)
end = datetime(2025, 1, 1)  

current = start
while current < end:
    date_start = current.strftime('%Y-%m-%d')
    date_end = (current + relativedelta(months=1)).strftime('%Y-%m-%d')
    
    print(f"Baixando dados de {date_start} até {date_end}")
    
    url = f"https://api.gridstatus.io/v1/datasets/ercot_standardized_hourly/query?api_key={api_key}&start_time={date_start}T00:00Z&end_time={date_end}T00:00Z"
    
    response = requests.get(url)
    if response.status_code == 200:
        df = pd.DataFrame(json.loads(response.text)['data'])
        df_total = pd.concat([df_total, df], ignore_index=True)
    else:
        print(f"Erro na requisição para {date_start} - Status code: {response.status_code}")
    
    current += relativedelta(months=1)

df_total

# df_total.to_excel("load_forecast.xlsx")
```
